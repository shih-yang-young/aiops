```python
import pandas as pd
df = pd.read_csv("../../data/kaggle_47837.csv")
df
```

```python
df["Inbound Message"] = df["Document"]
df["Assignment group"] = df["Topic_group"]
df.drop(["Document"], axis=1, inplace=True)
df.drop(["Topic_group"], axis=1, inplace=True)
df
```

```python
from preprocessing import clean_noise_text, preprocess_text
df['Inbound Message'] = df['Inbound Message'].apply(clean_noise_text)
df['Inbound Message'] = df['Inbound Message'].apply(preprocess_text)
df
```

```python
from preprocessing import remove_nan
df = remove_nan(df)
df
```

```python
df = df.drop_duplicates(subset=["Inbound Message"])
df
```

```python
from preprocessing import label_encode_column
df = label_encode_column(df.copy(), 'Assignment group', 'Label')
df = df.drop(['Assignment group'], axis=1)
df
```
