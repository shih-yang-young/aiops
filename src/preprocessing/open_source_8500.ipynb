```python
import pandas as pd
df = pd.read_excel("../../data/open_source_8500.xlsx") ## 舊名 IT-Ticket-Classification.xlsx
df
```

```python
df["Inbound Message"] = df["Short description"].astype(str) + " " + df["Description"].astype(str)
df.drop(['Short description'],axis=1,inplace= True)
df.drop(['Description'],axis=1,inplace= True)
df
```

```python
def replace_name(text, callers):
    if not callers or pd.isna(callers[0]):
        return text
    for name in callers:
        namelist = name.split()
        for namepart in namelist:
            text = text.replace(namepart, '')
            text = text.replace(namepart.lower(), '')
            text = text.replace(namepart.capitalize(), '')
    return ' '.join(text.split())

# 對每一列分別處理 Inbound Message + Caller
df['Inbound Message'] = df.apply(
    lambda row: replace_name(row['Inbound Message'], [row['Caller']]), axis=1
)

# 去掉 Caller 欄位
df.drop(["Caller"], axis=1, inplace=True)
df
```

```python
from preprocessing import clean_noise_text, preprocess_text
df['Inbound Message'] = df['Inbound Message'].apply(clean_noise_text)
df['Inbound Message'] = df['Inbound Message'].apply(preprocess_text)
df
```

```python
from preprocessing import remove_nan
df = remove_nan(df)
df
```

```python
df = df.drop_duplicates(subset=["Inbound Message"])
df
```

```python
## 將樣本數少於10的組別歸類為一個新的組別 misc_grp

# 計算各 Assignment group 的樣本數
group_counts = df['Assignment group'].value_counts()

# 取得樣本數少於 10 的 group 名稱
small_groups = group_counts[group_counts < 10].index

# 印出總共幾個 group 會被歸類為 misc_grp
print(f'Found {len(small_groups)} groups which have under 10 samples')

# 將這些 group 名稱改成 'misc_grp'
df.loc[:, 'Assignment group'] = df['Assignment group'].apply(
    lambda x: 'misc_grp' if x in small_groups else x
)
# Unique Groups check 
df['Assignment group'].unique()
```

```python
from preprocessing import label_encode_column
df = label_encode_column(df.copy(), 'Assignment group', 'Label')
# df = df.drop(['Assignment group'], axis=1)
df
```
