```python
import pandas as pd
open_source_8500 = pd.read_excel("../../data/open_source_8500.xlsx") ## 舊名 IT-Ticket-Classification.xlsx
open_source_8500
```

```python
open_source_8500["Inbound Message"] = open_source_8500["Short description"].astype(str) + " " + open_source_8500["Description"].astype(str)
open_source_8500.drop(['Short description'],axis=1,inplace= True)
open_source_8500.drop(['Description'],axis=1,inplace= True)
open_source_8500
```

```python
def replace_name(text, callers):
    if not callers or pd.isna(callers[0]):
        return text
    for name in callers:
        namelist = name.split()
        for namepart in namelist:
            text = text.replace(namepart, '')
            text = text.replace(namepart.lower(), '')
            text = text.replace(namepart.capitalize(), '')
    return ' '.join(text.split())

# 對每一列分別處理 Inbound Message + Caller
open_source_8500['Inbound Message'] = open_source_8500.apply(
    lambda row: replace_name(row['Inbound Message'], [row['Caller']]), axis=1
)

# 去掉 Caller 欄位
open_source_8500.drop(["Caller"], axis=1, inplace=True)
open_source_8500
```

```python
from my_preprocessing import clean_noise_text, preprocess_text
open_source_8500['Inbound Message'] = open_source_8500['Inbound Message'].apply(clean_noise_text)
open_source_8500['Inbound Message'] = open_source_8500['Inbound Message'].apply(preprocess_text)
open_source_8500
```

```python
from my_preprocessing import remove_nan
open_source_8500 = remove_nan(open_source_8500)
open_source_8500
```

```python
open_source_8500 = open_source_8500.drop_duplicates(subset=["Inbound Message"])
open_source_8500
```

```python
## 將樣本數少於10的組別歸類為一個新的組別 misc_grp

# 計算各 Assignment group 的樣本數
group_counts = open_source_8500['Assignment group'].value_counts()

# 取得樣本數少於 10 的 group 名稱
small_groups = group_counts[group_counts < 10].index

# 印出總共幾個 group 會被歸類為 misc_grp
print(f'Found {len(small_groups)} groups which have under 10 samples')

# 將這些 group 名稱改成 'misc_grp'
open_source_8500.loc[:, 'Assignment group'] = open_source_8500['Assignment group'].apply(
    lambda x: 'misc_grp' if x in small_groups else x
)
# Unique Groups check 
open_source_8500['Assignment group'].unique()
```
