```python
import pandas as pd
raw_data = pd.read_excel("dataset/IT-Ticket-Classification.xlsx")
raw_data
```

```python
# Count occurrences of each label in the Assignment group column
assignment_group_counts = raw_data["Assignment group"].value_counts()


# 將結果轉換為 DataFrame
assignment_group_counts_df = assignment_group_counts.reset_index()
assignment_group_counts_df.columns = ["Assignment Group", "Count"]

# 或者在 Jupyter Notebook 中顯示更清晰的表格
from IPython.display import display
assignment_group_counts_df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
plt.pie(
    assignment_group_counts_df["Count"],
    labels=assignment_group_counts_df["Assignment Group"],
    autopct='%1.1f%%',
    startangle=140
)
plt.title("Assignment Group Distribution")
plt.axis('equal')  # 保證圓餅圖為圓形
plt.show()
```

```python
kaggle_data = pd.read_csv("dataset/IT-Service-Ticket-Classification-Dataset.csv")
kaggle_data
```

```python
# Count occurrences of each label in the Assignment group column
kaggle_data_group_counts = kaggle_data["Topic_group"].value_counts()


# 將結果轉換為 DataFrame
kaggle_data_group_counts_df = kaggle_data_group_counts.reset_index()
kaggle_data_group_counts_df.columns = ["Topic_group", "Count"]

kaggle_data_group_counts_df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
plt.pie(
    kaggle_data_group_counts_df["Count"],
    labels=kaggle_data_group_counts_df["Topic_group"],
    autopct='%1.1f%%',
    startangle=140
)
plt.title("Assignment Group Distribution")
plt.axis('equal')  # 保證圓餅圖為圓形
plt.show()
```
