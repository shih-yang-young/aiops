```python
import pandas as pd
open_source_8500 = pd.read_excel("../../data/open_source_8500.xlsx")
open_source_8500
```

```python
label_counts = open_source_8500["Assignment group"].value_counts()
print(label_counts)
```

```python
# Count occurrences of each label in the Assignment group column
open_source_8500_counts = open_source_8500["Assignment group"].value_counts()


# 將結果轉換為 DataFrame
open_source_8500_counts_df = open_source_8500_counts.reset_index()
open_source_8500_counts_df.columns = ["Assignment Group", "Count"]

# 或者在 Jupyter Notebook 中顯示更清晰的表格
from IPython.display import display
open_source_8500_counts_df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
plt.pie(
    open_source_8500_counts_df["Count"],
    labels=open_source_8500_counts_df["Assignment Group"],
    autopct='%1.1f%%',
    startangle=140
)
plt.title("Assignment Group Distribution")
plt.axis('equal')  # 保證圓餅圖為圓形
plt.show()
```
